```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Upload the CSV file
uploaded = files.upload()   # upload netflix_titles.csv manually

# Load dataset
df = pd.read_csv("netflix_titles.csv")

# Basic checks
print(df.head())   # show first 5 rows
print("Shape of dataset:", df.shape)    # rows, columns
print("Columns:", df.columns)           # column names
print(df.info())                        # data types
print(df.isnull().sum())                # missing values

# Count of Movies vs TV Shows
print(df['type'].value_counts())
df['type'].value_counts().plot(kind='bar', color=['red', 'blue'])
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()

# Top 5 countries
top_countries = df['country'].value_counts().head(5)
print(top_countries)

top_countries.plot(kind='bar', color='green')
plt.title("Top 5 Countries with Most Netflix Content")
plt.show()

# --- FIX for date_added column ---
df['date_added'] = df['date_added'].astype(str).str.strip()  # remove extra spaces
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')  # convert safely

# Extract year
df['year_added'] = df['date_added'].dt.year

# Plot content added over years
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title("Content Added Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()

```
